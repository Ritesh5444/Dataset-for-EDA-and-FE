```python
#Taking country code dataset

df1 = pd.read_excel('https://github.com/Ritesh5444/Dataset-for-EDA-and-FE/raw/refs/heads/main/Country-Code.xlsx')
df1.head()
```

```python
df1.columns
```

```python
final_df = pd.merge(df,df1,on = 'Country Code',how = 'left')
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.shape
```

```python
# Now we will try to get some insight about the data

final_df['Country'].value_counts()
```

From above we can conclude that maximum number of Zomato restaurnats are present in India

```python
country_names = final_df['Country'].value_counts().index
country_names
```

```python
country_names.dtype
```

```python
country_value = final_df['Country'].value_counts().values
country_value
```

```python
#Creating a pie chart to see the market occupancy country wise

plt.pie(country_value [:3],labels = country_names[:3], autopct = '%1.2ff%%')
plt.show()
```

Conclusion - From above we can conclude that maximum number of Zomato restaurnats are present in India with value as 94.39%.
Followed by USA and then UK.

```python
#Trying to find the relation between Agreegate rating, Rating color and Rating text
```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size()
```

```python
Conclusion - It can be seen that rating color is related to Rating text and Aggregate rating value. With value range to each colour based on performance.
```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
rating = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
rating
```

```python
#Observation - 

1. When rating is between 4.5 to 4.9, the rating is Excellent
2. When rating is between 4.0 to 4.4, the rating is Very Good
3. When rating is between 3.5 to 3.9, the rating is Good
4. When rating is between 2.5 to 3.4, the rating is Average
5. When rating is between 1.8 to 2.4, the rating is Poor
6. Maximum people have not given rating in the app.
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = 15,6
```

```python

sns.barplot(data = rating,x= "Rating text", y = "Rating Count")
plt.show()
```

```python
sns.barplot(data = rating, x = "Aggregate rating", y = "Rating Count")
plt.show()
```

```python
sns.barplot(data = rating, x = "Aggregate rating", y = "Rating Count",hue = 'Rating color', palette = ['White','Red','Orange','Yellow','Green', 'Green'])
plt.show()
```

```python
#Observation
1. Not rated count is very high.
2. Maximum number of rating is between 2.7 to 3.9
```

```python
sns.countplot(data = rating, x = 'Rating color', palette = ['White','Red','Orange','Yellow','Green','Green'])
plt.show()
```
